```python
def preprocess_dataset(sequences, labels, max_length=None):
    '''
        Args:
            sequences: list, the list which contains the protein primary sequences.
            labels: list, the list which contains the dataset labels.
            max_length, Integer, the maximum sequence length,
            if there is a sequence that is larger than the specified sequence length will be post-truncated.
    '''
    if max_length is None:
        max_length = len(max(train_set.sequence.values, key=lambda x: len(x)))
    splitted_sequences = [list(seq[:max_length]) for seq in sequences]
    return splitted_sequences, labels
```

```python
def embed_dataset(model, sequences, shift_left = 0, shift_right = -1):
    inputs_embedding = []
    with torch.no_grad():
        for sample in tqdm(sequences):
            ids = tokenizer.batch_encode_plus([sample], add_special_tokens=True,
                                              padding=True, is_split_into_words=True,
                                              return_tensors="pt")
            embedding = model(input_ids=ids['input_ids'].to(device))[0]
            embedding = embedding[0].detach().cpu().numpy()[shift_left:shift_right]
            inputs_embedding.append(embedding)
    return inputs_embedding
```

```python
#@title Generate embeddings. { display-mode: "form" }
# Generate embeddings via batch-processing
# per_residue indicates that embeddings for each residue in a protein should be returned.
# per_protein indicates that embeddings for a whole protein should be returned (average-pooling)
# max_residues gives the upper limit of residues within one batch
# max_seq_len gives the upper sequences length for applying batch-processing
# max_batch gives the upper number of sequences per batch
def get_embeddings( model, tokenizer, seqs, per_residue, per_protein, sec_struct,
                   max_residues=4000, max_seq_len=1000, max_batch=100 ):

    if sec_struct:
      sec_struct_model = load_sec_struct_model()

    results = {"residue_embs" : dict(),
               "protein_embs" : dict(),
               "sec_structs" : dict()
               }

    # sort sequences according to length (reduces unnecessary padding --> speeds up embedding)
    seq_dict   = sorted( seqs.items(), key=lambda kv: len( seqs[kv[0]] ), reverse=True )
    start = time.time()
    batch = list()
    for seq_idx, (pdb_id, seq) in enumerate(seq_dict,1):
        seq = seq
        seq_len = len(seq)
        seq = ' '.join(list(seq))
        batch.append((pdb_id,seq,seq_len))

        # count residues in current batch and add the last sequence length to
        # avoid that batches with (n_res_batch > max_residues) get processed
        n_res_batch = sum([ s_len for  _, _, s_len in batch ]) + seq_len
        if len(batch) >= max_batch or n_res_batch>=max_residues or seq_idx==len(seq_dict) or seq_len>max_seq_len:
            pdb_ids, seqs, seq_lens = zip(*batch)
            batch = list()

            # add_special_tokens adds extra token at the end of each sequence
            token_encoding = tokenizer.batch_encode_plus(seqs, add_special_tokens=True, padding="longest")
            input_ids      = torch.tensor(token_encoding['input_ids']).to(device)
            attention_mask = torch.tensor(token_encoding['attention_mask']).to(device)

            try:
                with torch.no_grad():
                    # returns: ( batch-size x max_seq_len_in_minibatch x embedding_dim )
                    embedding_repr = model(input_ids, attention_mask=attention_mask)
            except RuntimeError:
                print("RuntimeError during embedding for {} (L={})".format(pdb_id, seq_len))
                continue

            if sec_struct: # in case you want to predict secondary structure from embeddings
              d3_Yhat, d8_Yhat, diso_Yhat = sec_struct_model(embedding_repr.last_hidden_state)


            for batch_idx, identifier in enumerate(pdb_ids): # for each protein in the current mini-batch
                s_len = seq_lens[batch_idx]
                # slice off padding --> batch-size x seq_len x embedding_dim
                emb = embedding_repr.last_hidden_state[batch_idx,:s_len]
                if sec_struct: # get classification results
                    results["sec_structs"][identifier] = torch.max( d3_Yhat[batch_idx,:s_len], dim=1 )[1].detach().cpu().numpy().squeeze()
                if per_residue: # store per-residue embeddings (Lx1024)
                    results["residue_embs"][ identifier ] = emb.detach().cpu().numpy().squeeze()
                if per_protein: # apply average-pooling to derive per-protein embeddings (1024-d)
                    protein_emb = emb.mean(dim=0)
                    results["protein_embs"][identifier] = protein_emb.detach().cpu().numpy().squeeze()


    passed_time=time.time()-start
    avg_time = passed_time/len(results["residue_embs"]) if per_residue else passed_time/len(results["protein_embs"])
    print('\n############# EMBEDDING STATS #############')
    print('Total number of per-residue embeddings: {}'.format(len(results["residue_embs"])))
    print('Total number of per-protein embeddings: {}'.format(len(results["protein_embs"])))
    print("Time for generating embeddings: {:.1f}[m] ({:.3f}[s/protein])".format(
        passed_time/60, avg_time ))
    print('\n############# END #############')
    return results
```

```python
# Add this section after generating embeddings and before training the model
# t-SNE Analysis of ANKH Embeddings

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
import numpy as np
import pandas as pd
from collections import Counter
import json # Import json

# Set up visualization parameters
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['font.size'] = 12
sns.set_style("whitegrid")

def analyze_embeddings_distribution(embeddings, labels, dataset_name="Dataset"):
    """
    Analyze the distribution and statistics of embeddings
    """
    print(f"\n========== {dataset_name} Embedding Analysis ==========")

    # Convert per-residue embeddings to per-protein embeddings by taking the mean
    # This ensures fixed-size vectors for further analysis
    embeddings_array = np.array([
        emb.mean(axis=0) if len(emb.shape) > 1 else emb
        for emb in embeddings
    ])

    print(f"Embedding shape: {embeddings_array.shape}")
    print(f"Embedding dimension: {embeddings_array.shape[1]}")
    print(f"Number of samples: {embeddings_array.shape[0]}")

    # Label distribution
    label_counts = Counter(labels)
    print(f"Label distribution: {dict(label_counts)}")

    # Embedding statistics
    print(f"Embedding mean: {np.mean(embeddings_array):.4f}")
    print(f"Embedding std: {np.std(embeddings_array):.4f}")
    print(f"Embedding min: {np.min(embeddings_array):.4f}")
    print(f"Embedding max: {np.max(embeddings_array):.4f}")

    return embeddings_array

def perform_tsne_analysis(embeddings, labels, dataset_name="Dataset",
                         perplexity=30, n_iter=1000, random_state=42):
    """
    Perform t-SNE analysis on embeddings and create visualizations
    """
    print(f"\n========== t-SNE Analysis for {dataset_name} ==========")

    # Convert to numpy array if needed
    # This part is already handling the conversion to per-protein if needed,
    # but analyze_embeddings_distribution will now return per-protein arrays,
    # so this check becomes less critical for correctness but still safe.
    if isinstance(embeddings, list):
        embeddings_array = np.array([
            emb.mean(axis=0) if len(emb.shape) > 1 else emb
            for emb in embeddings
        ])
    else:
        embeddings_array = embeddings

    labels_array = np.array(labels)

    print(f"Input shape for t-SNE: {embeddings_array.shape}")

    # Reduce dimensionality first with PCA if embeddings are very high-dimensional
    if embeddings_array.shape[1] > 50:
        print("Applying PCA preprocessing for t-SNE...")
        pca = PCA(n_components=50, random_state=random_state)
        embeddings_pca = pca.fit_transform(embeddings_array)
        print(f"PCA explained variance ratio: {pca.explained_variance_ratio_[:5]}")
        print(f"Total variance explained by first 50 components: {pca.explained_variance_ratio_.sum():.3f}")
    else:
        embeddings_pca = embeddings_array

    # Perform t-SNE
    print("Performing t-SNE...")
    tsne = TSNE(
        n_components=2,
        perplexity=min(perplexity, len(embeddings_array)//4),  # Adjust perplexity if needed
        n_iter=n_iter,
        random_state=random_state,
        verbose=1
    )

    embeddings_tsne = tsne.fit_transform(embeddings_pca)

    # Return the tsne object along with the results
    return embeddings_tsne, labels_array, tsne

def plot_tsne_results(embeddings_tsne, labels, dataset_name="Dataset", save_path=None, tsne_obj=None):
    """
    Create comprehensive t-SNE visualizations
    """
    # Create a DataFrame for easier plotting
    df_tsne = pd.DataFrame({
        'TSNE-1': embeddings_tsne[:, 0],
        'TSNE-2': embeddings_tsne[:, 1],
        'Label': labels,
        'Class': ['DBP' if label == 1 else 'Non-DBP' for label in labels]
    })

    # Create subplots
    fig, axes = plt.subplots(2, 2, figsize=(16, 12))
    fig.suptitle(f't-SNE Analysis of ANKH Embeddings - {dataset_name}', fontsize=16, fontweight='bold')

    # Plot 1: Basic scatter plot
    ax1 = axes[0, 0]
    colors = ['#FF6B6B', '#4ECDC4']  # Red for Non-DBP, Teal for DBP
    for i, (label, color, class_name) in enumerate(zip([0, 1], colors, ['Non-DBP', 'DBP'])):
        mask = df_tsne['Label'] == label
        ax1.scatter(df_tsne.loc[mask, 'TSNE-1'], df_tsne.loc[mask, 'TSNE-2'],
                   c=color, label=f'{class_name} (n={mask.sum()})',
                   alpha=0.7, s=50)

    ax1.set_xlabel('t-SNE Component 1')
    ax1.set_ylabel('t-SNE Component 2')
    ax1.set_title('t-SNE Visualization by Class')
    ax1.legend()
    ax1.grid(True, alpha=0.3)

    # Plot 2: Density plot
    ax2 = axes[0, 1]
    for i, (label, color, class_name) in enumerate(zip([0, 1], colors, ['Non-DBP', 'DBP'])):
        mask = df_tsne['Label'] == label
        ax2.scatter(df_tsne.loc[mask, 'TSNE-1'], df_tsne.loc[mask, 'TSNE-2'],
                   c=color, alpha=0.6, s=30)

    # Add density contours
    from scipy.stats import gaussian_kde
    for i, (label, color) in enumerate(zip([0, 1], colors)):
        mask = df_tsne['Label'] == label
        if mask.sum() > 1:
            data = df_tsne.loc[mask, ['TSNE-1', 'TSNE-2']].values.T
            kde = gaussian_kde(data)

            # Create a grid for contour plotting
            x_min, x_max = df_tsne['TSNE-1'].min(), df_tsne['TSNE-1'].max()
            y_min, y_max = df_tsne['TSNE-2'].min(), df_tsne['TSNE-2'].max()
            xx, yy = np.mgrid[x_min:x_max:.5, y_min:y_max:.5]
            positions = np.vstack([xx.ravel(), yy.ravel()])
            f = np.reshape(kde(positions).T, xx.shape)

            ax2.contour(xx, yy, f, colors=color, alpha=0.5, linewidths=1)

    ax2.set_xlabel('t-SNE Component 1')
    ax2.set_ylabel('t-SNE Component 2')
    ax2.set_title('t-SNE with Density Contours')
    ax2.grid(True, alpha=0.3)

    # Plot 3: Histogram of distances
    ax3 = axes[1, 0]

    # Calculate distances from centroid for each class
    centroids = {}
    distances = {}

    for label, class_name in zip([0, 1], ['Non-DBP', 'DBP']):
        mask = df_tsne['Label'] == label
        if mask.sum() > 0:
            class_data = df_tsne.loc[mask, ['TSNE-1', 'TSNE-2']].values
            centroid = np.mean(class_data, axis=0)
            centroids[class_name] = centroid

            # Calculate distances from centroid
            dist = np.sqrt(np.sum((class_data - centroid)**2, axis=1))
            distances[class_name] = dist

    # Plot histograms
    colors_hist = ['#FF6B6B', '#4ECDC4']
    for i, (class_name, color) in enumerate(zip(['Non-DBP', 'DBP'], colors_hist)):
        if class_name in distances:
            ax3.hist(distances[class_name], bins=20, alpha=0.7,
                    color=color, label=f'{class_name}', density=True)

    ax3.set_xlabel('Distance from Class Centroid')
    ax3.set_ylabel('Density')
    ax3.set_title('Distribution of Distances from Class Centroids')
    ax3.legend()
    ax3.grid(True, alpha=0.3)

    # Plot 4: Statistics summary
    ax4 = axes[1, 1]
    ax4.axis('off')

    # Calculate separation metrics
    if len(centroids) == 2:
        centroid_distance = np.sqrt(np.sum((centroids['DBP'] - centroids['Non-DBP'])**2))

        # Calculate average intra-class distances
        avg_intra_distances = {}
        for class_name in distances:
            avg_intra_distances[class_name] = np.mean(distances[class_name])

        # Calculate silhouette-like metric (simplified)
        dbp_mask = df_tsne['Label'] == 1
        non_dbp_mask = df_tsne['Label'] == 0

        # Get t-SNE parameters safely
        perplexity_val = tsne_obj.perplexity if tsne_obj and hasattr(tsne_obj, 'perplexity') else 'N/A'
        n_iter_val = tsne_obj.n_iter if tsne_obj and hasattr(tsne_obj, 'n_iter') else 'N/A'
        kl_divergence_val = tsne_obj.kl_divergence_ if tsne_obj and hasattr(tsne_obj, 'kl_divergence_') else 'N/A'

        # Corrected f-string formatting for KL Divergence
        kl_divergence_str = f"{kl_divergence_val:.2f}" if isinstance(kl_divergence_val, float) else str(kl_divergence_val)

        stats_text = f"""
    Clustering Analysis:

    Class Distribution:
    • DBP: {dbp_mask.sum()} samples
    • Non-DBP: {non_dbp_mask.sum()} samples

    Separation Metrics:
    • Distance between centroids: {centroid_distance:.2f}
    • Avg intra-class distance (DBP): {avg_intra_distances.get('DBP', 0):.2f}
    • Avg intra-class distance (Non-DBP): {avg_intra_distances.get('Non-DBP', 0):.2f}

    t-SNE Parameters:
    • Perplexity: {perplexity_val}
    • Iterations: {n_iter_val}
    • KL Divergence: {kl_divergence_str}

    Interpretation:
    • Higher centroid distance = better separation
    • Lower intra-class distances = tighter clusters
    • Good separation if centroid distance >
      average intra-class distances
        """

        ax4.text(0.05, 0.95, stats_text, transform=ax4.transAxes,
                fontsize=10, verticalalignment='top', fontfamily='monospace',
                bbox=dict(boxstyle='round', facecolor='lightgray', alpha=0.8))

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"t-SNE plot saved to: {save_path}")

    plt.show()

    return df_tsne

def calculate_embedding_quality_metrics(embeddings_tsne, labels):
    """
    Calculate quantitative metrics for embedding quality
    """
    from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

    print("\n========== Embedding Quality Metrics ==========")

    # Silhouette Score (higher is better, range: -1 to 1)
    sil_score = silhouette_score(embeddings_tsne, labels)
    print(f"Silhouette Score: {sil_score:.4f}")
    print("  • Measures how similar samples are to their own cluster vs. other clusters")
    print("  • Range: -1 (worst) to 1 (best)")

    # Calinski-Harabasz Score (higher is better)
    ch_score = calinski_harabasz_score(embeddings_tsne, labels)
    print(f"Calinski-Harabasz Score: {ch_score:.4f}")
    print("  • Ratio of between-cluster to within-cluster dispersion")
    print("  • Higher values indicate better clustering")

    # Davies-Bouldin Score (lower is better)
    db_score = davies_bouldin_score(embeddings_tsne, labels)
    print(f"Davies-Bouldin Score: {db_score:.4f}")
    print("  • Average similarity between each cluster and its most similar cluster")
    print("  • Lower values indicate better clustering")

    return {
        'silhouette_score': sil_score,
        'calinski_harabasz_score': ch_score,
        'davies_bouldin_score': db_score
    }

# Helper function to convert numpy types to Python types for JSON serialization
def convert_numpy_to_python(obj):
    if isinstance(obj, np.floating):
        return float(obj)
    elif isinstance(obj, np.integer):
        return int(obj)
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, dict):
        return {convert_numpy_to_python(k): convert_numpy_to_python(v) for k, v in obj.items()} # Convert keys as well
    elif isinstance(obj, list):
        return [convert_numpy_to_python(item) for item in obj]
    else:
        return obj


# ========== MAIN ANALYSIS CODE ==========
# Add this after generating embeddings but before training

print("Starting Comprehensive ANKH Embedding Analysis...")

# Set up visualization directory
import os
# import json # Already imported above

# Create visualization directory (similar to the one created later in the original code)
viz_dir = f"/gdrive/MyDrive/visualizations_ankh_analysis"
os.makedirs(viz_dir, exist_ok=True)
print(f"Visualization directory created at: {viz_dir}")

# ========== ANKH EMBEDDING PROPERTIES ANALYSIS ==========

# First, analyze what makes ANKH embeddings effective for DBP prediction
print("Analyzing ANKH embedding properties and their relevance to DBP prediction...")

train_analysis_results = analyze_ankh_embedding_properties(
    training_embeddings, training_labels, training_sequences, "Training Set"
)

test_analysis_results = analyze_ankh_embedding_properties(
    test_embeddings, test_labels, test_sequences, "Test Set"
)

# Analyze biochemical relevance
analyze_biochemical_relevance(
    training_embeddings, training_labels, training_sequences, "Training Set"
)

# Create comprehensive interpretation plots
create_embedding_interpretation_plots(
    training_embeddings, training_labels, training_sequences,
    train_analysis_results,
    save_dir=viz_dir if 'viz_dir' in globals() else None
)

# ========== STANDARD EMBEDDING DISTRIBUTION ANALYSIS ==========

# Analyze embedding distributions
train_embeddings_array = analyze_embeddings_distribution(
    training_embeddings, training_labels, "Training Set"
)

test_embeddings_array = analyze_embeddings_distribution(
    test_embeddings, test_labels, "Test Set"
)

# Perform t-SNE analysis on training set
print("\nPerforming t-SNE analysis on training embeddings...")
train_tsne, train_labels_array, train_tsne_obj = perform_tsne_analysis(
    training_embeddings, training_labels, "Training Set"
)

# Create visualizations for training set
train_tsne_df = plot_tsne_results(
    train_tsne, train_labels_array, "Training Set",
    save_path=f"{viz_dir}/tsne_training_embeddings.png" if 'viz_dir' in globals() else None,
    tsne_obj=train_tsne_obj # Pass the tsne object
)

# Calculate quality metrics for training embeddings
train_quality_metrics = calculate_embedding_quality_metrics(train_tsne, train_labels_array)

# Perform t-SNE analysis on test set
print("\nPerforming t-SNE analysis on test embeddings...")
test_tsne, test_labels_array, test_tsne_obj = perform_tsne_analysis(
    test_embeddings, test_labels, "Test Set"
)

# Create visualizations for test set
test_tsne_df = plot_tsne_results(
    test_tsne, test_labels_array, "Test Set",
    save_path=f"{viz_dir}/tsne_test_embeddings.png" if 'viz_dir' in globals() else None,
    tsne_obj=test_tsne_obj # Pass the tsne object
)

# Calculate quality metrics for test embeddings
test_quality_metrics = calculate_embedding_quality_metrics(test_tsne, test_labels_array)

# Combined analysis
print("\n========== Combined Dataset Analysis ==========")

# Combine training and test data for overall visualization
combined_embeddings = training_embeddings + test_embeddings
combined_labels = list(training_labels) + list(test_labels)
combined_dataset_labels = ['Train'] * len(training_labels) + ['Test'] * len(test_labels)

combined_tsne, combined_labels_array, combined_tsne_obj = perform_tsne_analysis(
    combined_embeddings, combined_labels, "Combined Dataset"
)

# Create a more detailed combined visualization
df_combined = pd.DataFrame({
    'TSNE-1': combined_tsne[:, 0],
    'TSNE-2': combined_tsne[:, 1],
    'Label': combined_labels_array,
    'Dataset': combined_dataset_labels[:len(combined_labels_array)],
    'Class': ['DBP' if label == 1 else 'Non-DBP' for label in combined_labels_array]
})

# Plot combined dataset with different markers for train/test
fig, axes = plt.subplots(1, 2, figsize=(16, 6))
fig.suptitle('Combined Dataset t-SNE Analysis', fontsize=16, fontweight='bold')

# Plot by class
ax1 = axes[0]
colors = ['#FF6B6B', '#4ECDC4']
for i, (label, color, class_name) in enumerate(zip([0, 1], colors, ['Non-DBP', 'DBP'])):
    mask = df_combined['Label'] == label
    ax1.scatter(df_combined.loc[mask, 'TSNE-1'], df_combined.loc[mask, 'TSNE-2'],
               c=color, label=f'{class_name} (n={mask.sum()})',
               alpha=0.7, s=50)

ax1.set_xlabel('t-SNE Component 1')
ax1.set_ylabel('t-SNE Component 2')
ax1.set_title('By Class')
ax1.legend()
ax1.grid(True, alpha=0.3)

# Plot by dataset
ax2 = axes[1]
markers = ['o', '^']  # Circle for train, triangle for test
for dataset, marker in zip(['Train', 'Test'], markers):
    for i, (label, color, class_name) in enumerate(zip([0, 1], colors, ['Non-DBP', 'DBP'])):
        mask = (df_combined['Label'] == label) & (df_combined['Dataset'] == dataset)
        if mask.sum() > 0:
            ax2.scatter(df_combined.loc[mask, 'TSNE-1'], df_combined.loc[mask, 'TSNE-2'],
                       c=color, marker=marker,
                       label=f'{dataset} {class_name} (n={mask.sum()})',
                       alpha=0.7, s=50)

ax2.set_xlabel('t-SNE Component 1')
ax2.set_ylabel('t-SNE Component 2')
ax2.set_title('By Dataset and Class')
ax2.legend()
ax2.grid(True, alpha=0.3)

plt.tight_layout()
if 'viz_dir' in globals():
    plt.savefig(f"{viz_dir}/tsne_combined_analysis.png", dpi=300, bbox_inches='tight')
    print(f"Combined t-SNE plot saved to: {viz_dir}/tsne_combined_analysis.png")
plt.show()

# Save all metrics to a comprehensive report
if 'viz_dir' in globals():
    embedding_analysis_report = {
        'training_quality_metrics': train_quality_metrics,
        'test_quality_metrics': test_quality_metrics,
        'dataset_info': {
            'train_samples': len(training_labels),
            'test_samples': len(test_labels),
            'embedding_dimension': train_embeddings_array.shape[1],
            'train_class_distribution': dict(Counter(training_labels)),
            'test_class_distribution': dict(Counter(test_labels))
        }
    }

    # Convert numpy types to python types before saving
    embedding_analysis_report_python = convert_numpy_to_python(embedding_analysis_report)

    with open(f"{viz_dir}/embedding_analysis_report.json", 'w') as f:
        json.dump(embedding_analysis_report_python, f, indent=4)
    print(f"Embedding analysis report saved to: {viz_dir}/embedding_analysis_report.json")

print("\n========== Analysis Complete ==========")
print("\nKey Insights from ANKH Embedding Analysis:")
print("\n1. EMBEDDING PROPERTIES:")
print("   • ANKH creates fixed-size representations regardless of protein length")
print("   • High-dimensional embeddings (1536D) capture complex protein relationships")
print("   • Embeddings encode evolutionary and functional information")

print("\n2. DBP-SPECIFIC ADVANTAGES:")
print("   • DNA-binding proteins have conserved structural motifs")
print("   • ANKH captures these patterns through evolutionary training")
print("   • Positive charge distribution important for DNA binding is encoded")
print("   • Context-aware representation learning captures motif positioning")

print("\n3. VISUALIZATION INSIGHTS:")
print("   • Check t-SNE plots for distinct cluster formation")
print("   • Higher silhouette scores indicate better class separation")
print("   • Overlapping regions show challenging samples for classification")
print("   • Compare train/test distributions to check for dataset shift")

print("\n4. CLASSIFICATION IMPLICATIONS:")
print("   • Good embedding separation suggests high classification potential")
print("   • Feature importance shows which dimensions are most discriminative")
print("   • Biochemical relevance validates the biological meaning of embeddings")

print("\n5. WHY ANKH WORKS FOR DBP PREDICTION:")
print("   ✓ Captures evolutionary conservation patterns")
print("   ✓ Encodes structural and functional motifs")
print("   ✓ Context-aware sequence representation")
print("   ✓ Learned from millions of protein sequences")
print("   ✓ Implicitly understands protein family relationships")
```

```python
class SolubilityDataset(Dataset):
    def __init__(self, sequences, labels):
        self.sequences = sequences
        self.labels = labels

    def __getitem__(self, idx):
        embedding = self.sequences[idx]
        label = self.labels[idx]
        return {'embed':torch.tensor(embedding), 'labels': torch.tensor(label, dtype=torch.float32).unsqueeze(-1)}

    def __len__(self):
        return len(self.sequences)
```

```python
def model_init(embed_dim):
    hidden_dim = int(embed_dim / 2)
    num_hidden_layers = 1 # Number of hidden layers in ConvBert.
    nlayers = 1 # Number of ConvBert layers.
    nhead = 4
    dropout = 0.1
    conv_kernel_size = 7
    pooling = 'max' # available pooling methods ['avg', 'max']
    downstream_model = ankh.ConvBertForBinaryClassification(input_dim=embed_dim,
                                                            nhead=nhead,
                                                            hidden_dim=hidden_dim,
                                                            num_hidden_layers=num_hidden_layers,
                                                            num_layers=nlayers,
                                                            kernel_size=conv_kernel_size,
                                                            dropout=dropout,
                                                            pooling=pooling)
    return downstream_model.cuda()
```

```python
def compute_metrics(p: EvalPrediction, plot_confusion_matrix=False, plot_roc_curve=False):
    # Get raw probabilities before thresholding
    probs = torch.sigmoid(torch.tensor(p.predictions)).numpy()
    # Binary predictions using 0.5 threshold
    preds = (probs > 0.5).tolist()
    labels = p.label_ids.tolist()

    # Confusion matrix values
    cm = metrics.confusion_matrix(labels, preds)
    tn, fp, fn, tp = cm.ravel()

    # Calculating sensitivity and specificity
    sensitivity = tp / (tp + fn) if (tp + fn) != 0 else 0
    specificity = tn / (tn + fp) if (tn + fp) != 0 else 0

    # Calculate ROC curve points and AUC
    fpr, tpr, thresholds = metrics.roc_curve(labels, probs)
    auc = metrics.auc(fpr, tpr)

    # Plot ROC curve only if explicitly requested
    if plot_roc_curve:
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC curve (AUC = {auc:.3f})')
        plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title('Receiver Operating Characteristic (ROC) Curve')
        plt.legend(loc="lower right")
        plt.grid(True, linestyle='--', alpha=0.7)
        plt.show()

    # Plot confusion matrix only if explicitly requested
    if plot_confusion_matrix:
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", cbar=False,
                    annot_kws={"size": 20},  # Increase font size for annotations
                    xticklabels=["Predicted Non-DBP", "Predicted DBP"],
                    yticklabels=["Actual Non-DBP", "Actual DBP"])
        plt.title("Confusion Matrix")
        plt.xlabel("Predicted Labels")
        plt.ylabel("True Labels")
        plt.show()

    return {
        "accuracy": metrics.accuracy_score(labels, preds),
        "precision": metrics.precision_score(labels, preds),
        "recall": metrics.recall_score(labels, preds),  # Same as sensitivity
        "f1": metrics.f1_score(labels, preds),
        "MCC": metrics.matthews_corrcoef(labels, preds),
        "sensitivity": sensitivity,
        "specificity": specificity,
        "auc": auc,
        "confusion_matrix": cm.tolist(),
    }
```
